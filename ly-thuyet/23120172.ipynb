## Tiền xử lý

```python
# Cột "Started on" và "Completed on": chuyển sang kiểu dữ liệu datetime

new_data = {}

for header, values in data.items():
    if header in ['Started on', 'Completed']:
        new_value = []
        for val in values:
            val = dt.datetime.strptime(val, '%d/%m/%Y %H:%M')
            new_value.append(val)
        new_data[header] = new_value
```

```python
# Cột "Time taken": chuyển sang một con số thể hiện cho số phút

for header, values in data.items():
    if header == 'Time taken':
        new_value = []
        for val in values:
            day = 0
            hour = 0
            minute = 0
            second = 0
            temp_list = val.split(' ')
            for k in range(len(temp_list)):
                if 'day' in temp_list[k]:
                    day = int(temp_list[k-1])
                if 'hours' in temp_list[k] or 'hour' in temp_list[k]:
                    hour = int(temp_list[k-1])
                if 'mins' in temp_list[k]:
                    minute = int(temp_list[k-1])
                if 'secs' in temp_list[k]:
                    second = int(temp_list[k-1])
            val = day*24*60 + hour*60 + minute + second/60
            new_value.append(val)
        new_data[header] = new_value
```

```python
# Các cột "Grade/10.00", "Q. 1 /0.63", ...: chuyển sang kiểu dữ liệu số

for header, values in data.items():
    if '/' in header:
        new_value = []
        for val in values:
            val = float(val)
            new_value.append(val)
        new_data[header] = new_value 
        
for header, values in new_data.items():
    print(f'{header} {type(values[0])}')
```

## Tiền xử lý + phân tích dữ liệu để trả lời cho câu hỏi 1

*Câu hỏi 1: Cột điểm tổng “Grade/10.00” có phân bố các giá trị như thế nào?*

Ở đây không cần tiền xử lý mà sẽ đi luôn vào phân tích dữ liệu.

Đầu tiên, ta sẽ xem thử median của cột "Grade/10.00".

```python
grades = new_data['Grade/10.00']
sorted_grades = sorted(grades)
p = 50
i = int(p/100 * (len(sorted_grades)-1))
sorted_grades[i]
```

Để cảm thụ rõ hơn về phân bố các giá trị của cột "Grade/10.00", ta sẽ chia miền giá trị của cột này thành các khoảng (bin) và xem thử mỗi khoảng có bao nhiêu sinh viên.

Giả sử ta sẽ chia đoạn [0, 10] thành 5 bin như sau: [0, 2), [2, 4), [4, 6), [6, 8),[8,10]. Ta có thể lưu kết quả vào một dictionary, trong đó: key là một tuple cho biết 2 đầu của một bin (thay tuple bằng list thì có được không?), value là số lượng sinh viên rơi vào bin này.

***Sinh viên hoàn thành tiếp phần này như bài tập về nhà***

```python
#TODO (CODE HERE)
bins = [(0, 2), (2, 4), (4, 6), (6, 8), (8, 10)]
bin_counts = {k: 0 for k in bins}
for g in grades:
    for k in bins:
        left, right = k
        if (left <= g < right) or (k == (8,10) and g == 10):
            bin_counts[k] += 1
            break
            
for key, count in bin_counts.items():
    print(f"{key}: {count}")
```

Thay `tuple` bằng `list` thì có được không? Trả lời trực tiếp vào cell này

Câu trả lời là **không**. Vì Python yêu cầu key trong dictionary phải là hashable (không thay đổi được).

Nhưng `list` lại **không hashable** nên không thể thay `tuple` bằng `list`.

Ta có thể lưu kết quả vào một dictionary, trong đó: key là tên của cột ứng với câu quiz, value là điểm trung bình của cột. Dictionary này được sort tăng dần theo value.

***Sinh viên hoàn thành tiếp phần này như bài tập về nhà***

```python
#TODO (CODE HERE)
avg_score = {}

for key, value in data.items():
    if key.startswith('Q.'):
        score = [float(x) for x in value if x.strip() != '']
        avg = sum(score)/len(score) if score else 0
        avg_score[key] = avg
        
avg_score = dict(sorted(avg_score.items(), key=lambda x: x[1]))
```

### Đơn giản
#### Bước 1: Tính thống kê cơ bản
- Tính thời gian trung bình của tất cả sinh viên
- Tính thời gian trung vị (median) - đây là mốc chia sinh viên thành 2 nhóm

#### Bước 2: Chia nhóm so sánh
- **Nhóm làm nhanh:** Sinh viên làm bài dưới thời gian trung vị
- **Nhóm làm chậm:** Sinh viên làm bài trên hoặc bằng thời gian trung vị
- Tính điểm trung bình của mỗi nhóm

#### Bước 3: Phân tích kết quả

**Có 3 tình huống có thể xảy ra:**

- **Nhóm làm nhanh có điểm cao hơn:** Có thể sinh viên giỏi nắm kiến thức vững nên làm nhanh và chính xác
- **Nhóm làm chậm có điểm cao hơn:** Có thể sinh viên cẩn thận, suy nghĩ kỹ sẽ đạt kết quả tốt hơn
- **Hai nhóm điểm tương đương:** Không có mối liên hệ rõ ràng giữa thời gian và điểm số

***Sinh viên hoàn thành tiếp trong các cell kế như bài tập về nhà***

```python
# TODO (CODE AND MARKDOWN)

# Bước 1.
avg_time = sum(new_data['Time taken'])/len(new_data['Time taken'])
print('Thời gian trung bình của tất cả sinh viên: ', avg_time)

times = new_data['Time taken']
sorted_times = sorted(times)
p = 50
i = i = int(p / 100 * (len(sorted_times) - 1))
time_median = sorted_times[i]
print('Thời gian trung vị: ', time_median)


# Bước 2.
list_score_lower = []
list_score_higher = []

for i in range(len(new_data["Time taken"])):
    if new_data["Time taken"][i] < time_median:
        list_score_lower.append(new_data["Grade/10.00"][i])
    elif new_data["Time taken"][i] >= time_median:
        list_score_higher.append(new_data["Grade/10.00"][i])
        
avg_score_lower = sum(list_score_lower)/len(list_score_lower)
avg_score_higher = sum(list_score_higher)/len(list_score_higher)
print('Điểm trung bình của nhóm làm nhanh: ', avg_score_lower)
print('Điểm trung bình của nhóm làm chậm: ', avg_score_higher)
```

## Câu hỏi 4
**Có sinh viên nào là điểm ngoại lệ (outliers) không?**\
**Mục đích:** Tìm những sinh viên có điểm số bất thường (quá cao hoặc quá thấp so với mức trung bình).

### Phương pháp 1: Sử dụng khoảng tứ phân vị (IQR - Interquartile Range)

#### Bước 1: Tính các giá trị thống kê
- Sắp xếp tất cả điểm số từ thấp đến cao
- Tính **Q1** (tứ phân vị thứ nhất): Điểm ở vị trí 25% (1/4 đầu dãy số)
- Tính **Q3** (tứ phân vị thứ ba): Điểm ở vị trí 75% (3/4 dãy số)
- Tính **IQR** = Q3 - Q1 (khoảng giữa)

**Ví dụ:** Có 100 sinh viên sau khi sắp xếp
- Q1 = điểm của sinh viên thứ 25 (ví dụ: 4.5)
- Q3 = điểm của sinh viên thứ 75 (ví dụ: 7.5)
- IQR = 7.5 - 4.5 = 3.0

#### Bước 2: Xác định ngưỡng outlier
- **Ngưỡng dưới** = Q1 - 1.5 × IQR
- **Ngưỡng trên** = Q3 + 1.5 × IQR

Tiếp ví dụ trên:
- Ngưỡng dưới = 4.5 - 1.5 × 3.0 = 0
- Ngưỡng trên = 7.5 + 1.5 × 3.0 = 12 (nhưng tối đa là 10)

#### Bước 3: Tìm outliers
- Sinh viên có điểm < ngưỡng dưới → **Outlier thấp** (điểm kém bất thường)
- Sinh viên có điểm > ngưỡng trên → **Outlier cao** (điểm xuất sắc bất thường)

***Sinh viên hoàn thành tiếp trong các cell kế như bài tập về nhà***

```python
# TODO (CODE AND MARKDOWN)
grades = [float(x) for x in data["Grade/10.00"]]
grades_sorted = sorted(grades)

def tinh_phanvi(values, p):
    k = (len(values) - 1) * p
    f = int(k)
    c = f + 1
    if c >= len(values):
        return values[f]
    return values[f] + (values[c] - values[f]) * (k - f)

Q1 = tinh_phanvi(grades_sorted, 0.25)
Q3 = tinh_phanvi(grades_sorted, 0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
    
outliers_low = [g for g in grades if g < lower_bound]
outliers_high = [g for g in grades if g > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Ngưỡng dưới:", lower_bound)
print("Ngưỡng trên:", upper_bound)
print("Outlier thấp:", outliers_low)
print("Outlier cao:", outliers_high)
```

`Kết luận`: Chỉ có 1 sinh viên điểm kém bất thường. Không có sinh viên nào có điểm xuất sắc bất thường.

### Phương pháp 2: Sử dụng độ lệch chuẩn (Standard Deviation)

#### Bước 1: Tính giá trị trung bình và độ lệch chuẩn
- Tính điểm trung bình (mean)
- Tính độ lệch chuẩn (standard deviation) - đo mức độ phân tán của điểm số

#### Bước 2: Xác định ngưỡng
- **Ngưỡng dưới** = Mean - 2 × Std Dev
- **Ngưỡng trên** = Mean + 2 × Std Dev

**Ví dụ:**
- Điểm trung bình = 6.0
- Độ lệch chuẩn = 1.5
- Ngưỡng dưới = 6.0 - 2 × 1.5 = 3.0
- Ngưỡng trên = 6.0 + 2 × 1.5 = 9.0

#### Bước 3: Tìm outliers
- Điểm < 3.0 hoặc > 9.0 là outliers

***Sinh viên hoàn thành tiếp trong các cell kế như bài tập về nhà***

```python
# TODO (CODE AND MARKDOWN)
grades = [float(x) for x in data["Grade/10.00"]]
n = len(grades)

mean = sum(grades) / n
print("Điểm trung bình:", mean)

variance_population = sum((x - mean) ** 2 for x in grades) / n
std_population = variance_population ** 0.5
print("Độ lệch chuẩn:", std_population)

lower_bound = mean - 2 * std_population
upper_bound = mean + 2 * std_population

outliers_low = [g for g in grades if g < lower_bound]
outliers_high = [g for g in grades if g > upper_bound]

print("Ngưỡng dưới:", lower_bound)
print("Ngưỡng trên:", upper_bound)
print("Outlier thấp:", outliers_low)
print("Outlier cao:", outliers_high)
```

`Kết luận`: 3 sinh viên điểm kém bất thường. Không có sinh viên nào có điểm xuất sắc bất thường.
